```python
import numpy as np
import pandas as pd

n=200
x = np.linspace(0, 4*np.pi, n)
```

```python
s = pd.Series(np.sin(x), index=x)
s.plot()
```

```python
snoise = s + 0.1 * np.random.randn(n)
sdf = pd.DataFrame({'sin wave':s, 'noise wave': snoise})
sdf.plot(color=('r', 'b'))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(snoise)
```

```python
y = snoise
z=np.array(y).reshape(-1, 1)
print(len(z))
```

```python
x = list(range(0, 200, 1))
```

```python
w=np.array(x).reshape(-1, 1)
print(w)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(w, y)
regr_2.fit(w, y)
```

```python
# Predict
X_test = np.arange(0.0, 200, 1)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

```python
# Plot the results
plt.figure()
plt.scatter(w, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
